```python
import matplotlib.pyplot as plt

# pyplot is a collection of functions that matplotlib work as MATLAB
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 35, 45]

plt.plot(x, y) # line chart
plt.show() # display chart in jupyter notebook
```

##### customize chart

```python
plt.figure(figsize=(4, 3)) # figure size
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)

plt.title("This is a line chart")
plt.xlabel("X-AXIS")
plt.ylabel("Y-LABEL")
plt.show()
```

##### Advanced line chart - multiple lines and legends

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.plot(x, y1, label='Sales 2024')
plt.plot(x, y2, label='Sales 2025')

plt.title('YoY Sales')
plt.xlabel('Month')
plt.legend()
plt.show()
```
