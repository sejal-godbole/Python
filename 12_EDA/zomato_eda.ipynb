```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on = "Country Code", how = "left")
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels = country_names[:3], autopct="%1.2f%%")
```

Observations: zomato maximum records ot transactions are from India after that USA then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##### Conclusions
1. When the rating are between 4.5 to 4.9 -----> Excellent
2. When the rating are between 4.0 to 3.4 -----> It is very good
3. When the rating are between 3.5 to 3.9 -----> good
4. When the rating are between 3.0 to 3.4 -----> average
5. When the rating are between 2.5 to 2.9 -----> average
6. When the rating are between 2.0 to 2.4 -----> poor


```python
ratings.head()
```

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue="Rating color", palette=['white', 'red', 'orange', 'yellow', 'green', 'green'], edgecolor="black")
```

Observations
- Not rated count is very high
- Maximum number of ratings is between 2.5-3-4

```python
sns.countplot(x="Rating color", data=ratings,hue="Rating color", palette=['white', 'red', 'orange', 'yellow', 'green', 'green'], edgecolor="black")
```
