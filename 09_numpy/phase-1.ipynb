## Numpy array and basics

```python
import numpy as np
```

Creating array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array: ", arr_2d)
```

List vs Numpy array

```python
py_list = [1, 2, 3]
print("Python list multiplication", py_list * 2)

np_array = np.array([1, 2, 3]) # element wise multiplication
print("Python array multiplication", np_array * 2)

import time

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time: ", time.time() - start)

start = time.time()
np_array = np.array(1000000) * 2
print("\n Numpy operation time: ", time.time() - start)
```

Creating array from scratch

```python
zeros = np.zeros((3, 4))
print("zeros array: \n", zeros)

ones = np.ones((2, 3))
print("ones array: \n", ones)

full = np.full((2, 2), 7)
print("full array: \n", full)

random = np.random.random((2, 3))
print("random array: \n", random)

sequence = np.arange(0, 10, 2)
print("sequence array: \n", sequence)
```

## Vector, Matric and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Matrix: \n", matrix)

tensor = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])
print("Tensor: \n", tensor)
```

Array properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Shape ", arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("Data Type ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3, 4))
print("\nReshaped array ", reshaped)

flattened = reshaped.flatten()
print("\nFattened array ", flattened)

raveled = reshaped.ravel()
print("\n revaled array ", raveled) # ravel returns instead of copy

# Transpose
transpose = reshaped.T
print("\n Transposed array\n", transpose)
```
